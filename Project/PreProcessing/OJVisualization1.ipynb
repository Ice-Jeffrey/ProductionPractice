```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from pyecharts.charts import Boxplot
from pyecharts import options as opts
```

```python
# 读入需要进行可视化的数据
data = pd.read_csv('F:\Codes\ProductionPractice\OutputData\ExcelData3.csv')
data.head()
```

```python
# 去掉提交数为0的数据
data = data[data['OJ提交数'] != 0]
data.head()
```

```python
# 描述性数据统计
data.loc[:, ['OJ正确数', 'OJ提交数', 'OJ准确率']].describe()
```

```python
# 进行y轴数据的准备
y_data = [
    data['OJ正确数'],
    data['OJ提交数']
]

title = '正样本oj正确数与提交数箱型图'

# 绘制箱型图
box = (
    Boxplot(
        init_opts=opts.InitOpts(page_title=title)
    ).
    add_xaxis(['OJ正确数', 'OJ提交数']).
    add_yaxis(series_name="数量", y_axis=Boxplot.prepare_data(y_data)).
    set_global_opts(
        title_opts=opts.TitleOpts(title=title),
        xaxis_opts=opts.AxisOpts(name='类别'),
        yaxis_opts=opts.AxisOpts(name='数量')
    )
)
box.render('./OutputData/oj1.html')
```
